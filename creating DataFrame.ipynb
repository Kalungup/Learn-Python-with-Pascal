```python
import pandas as pd
import numpy as np
```

```python

np.random.seed(0)
student_marks = np.random.normal(loc=69, scale=30, size=200000)
student_marks = np.clip(student_marks, 0, 100).astype(int)

students = ['student ' +str(i) for i in range (1,200001)]

marks = {'Student Name': students, 'marks': student_marks}
df = pd.DataFrame(marks)


 
```

```python
sample = df.sample(n=500,random_state = 0)

sample
```

```python
population_mean = df['marks'].mean()
sample_mean =sample['marks'].mean()

print(f"The population mean is {population_mean} wheres the sample mean is {sample_mean}")
```

```python
# Population variance
population_variance = df['marks'].var(ddof=0)

# Sample variance
sample_variance = sample['marks'].var(ddof=1)

print("Population Mean:", population_mean)
print("Sample Mean:", sample_mean)
print("Population Variance:", population_variance)
print("Sample Variance:", sample_variance)
```

```python
import scipy.stats as stats
```

```python
sample_data = sample['marks']

mean = np.mean(sample_data)
standard_error = stats.sem(sample_data)

degrees_freedom = len(sample_data) - 1


# Calculate 95% confidence interval for sample mean
confidence_level = 0.95
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, mean, standard_error)

print(f"95% Confidence Interval: {confidence_interval}")
```

```python
import matplotlib.pyplot as plt

# Taking multiple samples and calculating their means
# That is, take a random sample of 15 values. Repeat this process 100 times.
sample_means = [np.mean(df['marks'].sample(30)) for _ in range(100)]

# Plotting the distribution of sample means
plt.hist(sample_means, bins=20, edgecolor='black')
plt.xlabel('Sample Means')
plt.ylabel('Frequency')
plt.title('Distribution of Sample Means')
plt.show()
```
